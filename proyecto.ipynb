El cambio climático es consecuencia de una externalidad negativa global originada por las emisiones de gases de efecto invernadero (GEI) a la atmósfera, sin un costo económico explícito en la gran mayoría de los países. Se puede afirmar que las emisiones totales están dictadas en gran medida por la población y producción de cada país.

Los datos estadísticos y científicos han demostrado que los fenómenos climáticos se ven perturbados por la concentración de los gases de efecto invernadero, emitidos a la atmósfera por el hombre, producto de los patrones de consumo y producción concebidos por el modo de desarrollo económico actual.

El calentamiento del sistema climático es inequívoco, como es evidente de las observaciones de las temperaturas globales promedio de la atmósfera y los océanos, del derretimiento de la nieve y hielo generalizado y del aumento global del nivel medio del mar. Hay cambios en las condiciones ambientales en períodos cortos de tiempo, que amenazan la supervivencia de los ecosistemas y las especies, y por lo tanto la forma de vida de las sociedades dependientes de ellos.

Los impactos a consecuencia del cambio climático ya están ocurriendo y los posibles impactos futuros han sido identificados.

El rol de las ciudades en el cambio climático, en términos de mitigación de emisiones, transacciones de créditos de carbono y adaptación a los impactos locales y regionales, es un tema de creciente interés y preocupación.

La mayor parte del calentamiento global observado durante el siglo XX se debe muy probablemente (90% de confianza) al aumento en las 
concentraciones de gases de efecto invernadero causado por las sociedades humanas.

Los altos niveles de consumo de energía de las sociedades desarrolladas, producto de la irreflexión y de los intereses económicos, han impuesto a la humanidad costos sociales de gran envergadura que pueden afectar a todas las formas de vida en la Tierra. A pesar de lo inciertas que puedan resultar las proyecciones de los escenarios climáticos, los patrones actuales de consumo energético conducen al mundo por un sendero insustentable que vincula la problemática energética con el cambio climático.

La siguiente investigacion se realiza buscando responder la pregunta: Cómo reducir las emisiones de carbono?. Es por ello que se considera una serie de datos tomados de SINAMEC relacionados con el cambio climático en Costa Rica.

<h2 style="color:#8A2BE2"> ANALISIS Y CONCLUSIONES</h2>

Después de hacer una revisión, validación y análisis de los datos de consumo por sector y haciendo interpretacion de las graficas gf1 y gf3, podemos determinar que: en el año 2018 el sector transporte obtuvo un consumo superior a los 80 mil, seguido del sector industrial con un consumo muy cercano a los 40 mil y en tercer lugar el sector residencial con un consumo de 20 mil. El sector servicio consumió cercano a los 10 mil y los sectores público, agropecuaria y comercial tuvieron consumos menores a 5mil.

En el año 2010 se muestra el mismo comportamiento en cuanto al orden pero con algunas variantes en cuanto a consumo de algunos sectores. Los mas relevante es que el sector residencial en este año consumía 23 mil lo que refleja que con respecto a 2018 hubo una disminución de consumo mientras que transporte aumentó de 70 mil a 80 mil en el 2018.

Después de hacer una revisión, validación y análisis de los datos de consumo por tipo de fuente y haciendo interpretacion de las graficas gf2 y gf4, podemos determinar que: en el año 2018 la fuente mas consumida fue el diesel con 45mil, seguido de la electricidad con 36 mil y en tercer lugar la gasolina super con 22 mil muy seguida del diesel con 21 mil. La leña, el bagazo, el glp y get fuel consumieron entre 9 y 8 mil. Las otras fuentes obtuvieron consumos menores a 4 mil. 

En el año 2010  se muestra que el diesel y la electricidad conservan los mismo puestos solo que con menor consumo que en el 2018. La gasolina regular estaba en el trecer lugar con un consumo de 19 mil. La leña en el cuarto lugar con un consumo de 14 mil.El bagazo se consumía en 9 mil, un numero más alto que en el 2018. 

Resulta interesante observar la disminucion del consumo de gasolina regular  y el aumento de gasolina super, ademas una disminucion en el 2018 del uso de bagazo y leña como fuente de energía.


El desafío planteado por el cambio climático está establecido en la agenda pública a niveles multilaterales y nacionales. Desde la construcción del Convenio Marco sobre Cambio Climático (UNFCCC), a principios de la década de 1990, el enfoque dominante ha estado en la mitigación de emisiones de GEI.

Por lo que es recomendable:
1.Aumentar las fuentes de producción de electricidad con fuentes renovables y que impactan positivamente el ambiente como energía solar, hidraúlica o geotérmic, que son fuentes con las que se cuenta en el país.
2.Promover el uso de autos que usen combustibles diferentes al diesel o gasolinas como la electricidad u otras mas amigables.

3.Fomentar en los hogares el uso de fuentes solares de energía.

4.Concientizar y educar a la población desde edades tempranas para lograr mayor compromiso de la sociedad y por ende menos impactos negativos al ambiente logrando reducir la emisión de gases.

5.Ofrecer créditos de carbono para financiar proyectos de desarrollo más limpios (de infraestructura, de transporte, industrial o de actividad agroforestal).

5.Planificar en torno a ellos en infraestructura, producción, vivienda, agricultura y recursos renovables es clave.

6.Incentivar las inversiones en energías renovables mediante el pago al generador de energía renovable independiente por la energía que suministra a la red.

7.Modificar la legislación para que estas propuestas sean viales es prioridad, acompañada de una estrategia que realmente implique el compromiso y participación de la sociedad.

8.En el mercado energético explorar las posibilidades de desarrollar más energías renovables.
